```python

with open("hp.txt", "r") as _if, open("wc.txt", "w") as of:
    words = {}
    for l in _if:
        for w in l.split():
            w = w.lower()
            for char in w:
                if not char.isalnum():
                    w = w.replace(char,"")
            if w in words:
                words[w] += 1
            elif w!= "":
                words[w] = 1

    for k,v in words.items():
        of.write("{} {}\n".format(k,v))


```

```python
s = []
with open("wc.txt", "r") as _if, open("wc-sorted.txt", "w") as of:
    for l in _if:
        word, num = l.split()
        num = int(num)
        s.append((word, num))

    s = sorted(s, reverse=True, key=lambda i: i[1])

    for x in s:
        of.write("{} {}\n".format(x[0], x[1]))
```
